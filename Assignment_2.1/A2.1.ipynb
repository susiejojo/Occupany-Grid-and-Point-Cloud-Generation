```python
import sys
import numpy as np
import open3d as o3d
sys.path.insert(1, "../dataset")
from utils import readData, readPointCloud
```

```python
ground_truth = readData("../dataset/01.txt")
ground_truth = ground_truth[:77][:]
pc = []
for i in range(1):
    file = "../dataset/01/" + str(i).zfill(6) + ".bin"
    data = readPointCloud(file)
    valid = []
    for j in range(len(data)):
        dist = np.linalg.norm(data[j][:2])
        if dist <= 50 and dist >= 10:
            valid.append(data[j])
    pc.append(valid)
```

```python
def computePose(point_ind):
#     final_pose = []
    T = ground_truth[point_ind][:]
    T = np.reshape(T, (3,4))
    b = np.array([0,0,0,1])
    T = np.vstack ((T,b)) 
    print (T)
    return T
```

```python
# print (ground_truth.shape)
res = computePose(76)
# print(np.shape(res[0]))
```

```python
pcd_net = o3d.geometry.PointCloud()
for i in range(76):
    if (i<9):
        pcd_text = "00000"+str(i+1)+".bin"
    else:
        pcd_text = "0000"+str(i+1)+".bin"
    pcd_ip = readPointCloud("../dataset/01/"+pcd_text)
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(pcd_ip[:,:3])
#     print (np.asarray(pcd.points).shape)
    rotmat = pcd.get_rotation_matrix_from_zyx((np.pi/2,-np.pi/2,0))
    pcd = pcd.rotate(R=rotmat,center=pcd.get_center())
    pcd = pcd.transform(computePose(i))
    pcd_net += pcd
#     print (np.asarray(pcd.points))
# pcd = o3d.io.read_point_cloud(pcd_ip)
downpcd = pcd_net.voxel_down_sample(voxel_size=1)
print (type(pcd))
o3d.visualization.draw_geometries([downpcd])
```
